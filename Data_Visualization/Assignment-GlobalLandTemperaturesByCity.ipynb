# Line plot:

### 1.How has the avg temp in a city changed over the years

 - x-axis:Years & y-axis: Avg Temp

```python
df['dt'].dtype
df['dt']=pd.to_datetime(df['dt'],errors='coerce')
df['dt'].dtype
```

```python
import matplotlib.pyplot as plt

df['year'] = df['dt'].dt.year

x=[df["City"]=="New York"]['dfyear'].unique()
y=df[df['City']=='New York'].groupby("year")['AverageTemperature'].mean()
plt.plot(x,y)
plt.title('Avg Temp of New York city changed over years')
plt.xlabel('Years')
plt.ylabel('Avg Temp')
plt.show()
```

# Bar plot:

### 2.Which countries had the highest avg temperature in the dataset

 - x-aixs:Countries & y-axis:Avg Temp

```python
countries_avg_temp=df.groupby('Country')['AverageTemperature'].mean().reset_index()
top_10_countries=countries_avg_temp.sort_values(by='AverageTemperature',ascending=False).head(10)
avg_temp=df['AverageTemperature']
top_10_countries
```

```python
plt.figure(figsize=(15,5))
plt.bar(top_10_countries['Country'],top_10_countries['AverageTemperature'])
plt.title('Top 10 Countries with highest avg temperature')
plt.xlabel('Countries')
plt.ylabel('Avg Temp')
plt.show()
```

# Histogram:

### 3.What is the distribution of avg temperatures across all data

 - x-axis:Temp ranges & y-axis: Frequencies

```python
avg_temp=df['AverageTemperature'].dropna()
```

```python
plt.hist(avg_temp,bins=50)
plt.title('Distribution of Avg Temperatures')
plt.xlabel('Avg Temp')
plt.ylabel('Frequency')
plt.show()
```

# Scatter Plot:

### 4.Correlation b/w temperature & uncertainity in the datset

 - x-axis: Avg Temp & y-axis:Temp uncertainity

```python
avg_temp=df['AverageTemperature'].dropna()
avg_temp_un=df['AverageTemperatureUncertainty'].dropna()
plt.scatter(avg_temp,avg_temp_un,color='purple',alpha=0.2)
#plt.figure(figsize=(10,6))
plt.title('Relation b/w Temperature & Uncertainity')
plt.xlabel('Avg Temp')
plt.ylabel('Uncertainity')
plt.show()
```

# Box Plot:

### 5.What is the spread of temperatures for a specific country over a given period

 - x-axis:years/countries & y-axis:Avg Temp

```python
country_india=df[df["Country"]=="India"].dropna(subset='AverageTemperature')
start_year=1900
end_year=2000
df_filtered = df[(df['Country'] == "India") & (df['year'] >= start_year) & (df['year'] <= end_year)]
India_year_mean=df_filtered.groupby('year')['AverageTemperature'].mean()
plt.boxplot(India_year_mean)
plt.title('Avg Temperatures of India from 1900 to 2000')
plt.show()
```
