```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Sales Analysis:

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales=np.random.randint(10,100,12)
plt.plot(months, sales, marker='*',mec='b', linestyle='--', color='y',label="Sales")
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.title("Monthly sales Line Plot")
plt.show()
```

```python
sns.lineplot(x=months,y=sales,marker='o',color='y')
plt.title("Monthly sales Line Plot")
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.show()
```

```python
plt.hist(sales,bins=12,color='g', alpha=0.5)
plt.xlabel("Order Quantity")
plt.ylabel("Frequency")
plt.title("Order Quantity Distribution")
plt.show()
```

```python
sns.histplot(sales, bins=12, kde=True, color='g')
plt.xlabel("Order Quantity")
plt.ylabel("Frequency")
plt.title("Order Quantity Distribution")
plt.show()
```

# Customer Demographics:

```python
age=np.random.randint(1,70,10)
plt.hist(age, bins=10, color='purple')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Customers")
plt.show()
```

```python
sns.histplot(age, bins=10, kde=True, color='purple')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Customers")
plt.show()
```

```python
spending = np.random.randint(50, 500, 10)  # Random spending amounts

plt.scatter(x=age, y=spending, alpha=0.6, color='r')
plt.xlabel("Age")
plt.ylabel("Spending ($)")
plt.title("Customer Spending Patterns")
plt.show()
```

```python
spending = np.random.randint(50, 500, 10)  # Random spending amounts

sns.scatterplot(x=age, y=spending, alpha=0.6, color='r')
plt.xlabel("Age")
plt.ylabel("Spending ($)")
plt.title("Customer Spending Patterns")
plt.show()
```

# School Performance:

```python
subjects=['Maths','Eng','Phy','Che','Social','Tel','Hindi']
marks=np.random.randint(1,101,7)
plt.bar(subjects, marks, color=['blue', 'red', 'green', 'orange', 'purple','yellow','pink'])
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Students' Marks Across Subjects")
plt.show()
```

```python
sns.barplot(x=subjects, y=marks, palette=['blue', 'red', 'green', 'orange', 'purple','yellow','pink'])
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Students' Marks Across Subjects")
plt.show()
```

```python
plt.boxplot(marks, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks Distribution Across Subjects")
plt.show()
```

```python
sns.boxplot(data=marks, color='lightblue')
plt.title("Marks Distribution with Outliers")
plt.show()
```
