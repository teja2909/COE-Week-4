```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color='yellow',linestyle='--',marker='h',mfc='green',mec='red',linewidth=1)
plt.title('Simple line plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1000,5000,10000,15000,30000,70000]
plt.plot(x,y,color='green',linestyle='--',marker='H',mfc='black')
plt.title('Growth of savings account')
plt.xlabel('Months(1-6)')
plt.ylabel('Amount')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['beige','skyblue'],ec='black')
plt.title('Basic Bar Chart!')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,36]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='coral',ec='black')
plt.title('Age Distribution of customers')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.show()
```

```python
import plotly.express as px
px.histogram(x,y)
```

# Pie Chart

```python
lang=['py','java','cpp','js','ruby','kotlin']
percent=[40,20,30,50,10,5]
plt.pie(percent,labels=lang,autopct='%1.1f%%',explode=(0.2,0,0,0.1,0,0),shadow=True)
plt.title('Programming languages')
plt.show()
```

```python
import plotly.express as px
px.pie(values=percent,names=lang,title='Pie Chart',hole=0.1)
```

```python
x=[5,10,15,20,25]
y=[10,20,25,38,35]
plt.scatter(x,y,color='red',s=50)
plt.show()
```
