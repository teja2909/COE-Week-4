# Seaborn:

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips
```

```python
tips.head()
```

# HISTOGRAM:

```python
import matplotlib.pyplot as plt
sns.histplot(data=tips,x='total_bill',bins=20,kde=True,color='MediumPurple')
plt.title('Total Bill Distribution')
plt.show()
```

# KDE plot:

```python
sns.kdeplot(data=tips,x='tip',color='MediumPurple',marker='4')
plt.show()
```

# Bar plot:

```python
sns.barplot(data=tips,x='day',y='total_bill',palette=['beige','skyblue'])
plt.show()
```

# Box plot:

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette="coolwarm")
plt.show()
```

# Scatter plot:

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='deep')
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='o',color='MediumOrchid')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.show()
```

```python
correlation_matrix=tips.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation of heatmap')
plt.show()
```
