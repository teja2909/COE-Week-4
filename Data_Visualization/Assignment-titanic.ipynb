```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('train.csv')
df
```

```python
df.head()
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',palette=['LightSalmon','pink',],hue='Sex')
plt.title('Survival rate by passenger class')
plt.xlabel('Passenger class')
plt.ylabel('Survival class')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',palette='coolwarm',alpha=0.5)
plt.title('Age vs Fare')
plt.show()
```

```python
sns.pairplot(data=df,vars=['Age','Fare','Pclass','Survived'],hue='Sex',diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,fmt='.2f',cmap='crest')
plt.show()
```
